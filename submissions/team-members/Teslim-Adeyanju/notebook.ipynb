````python
# Memory Management Utilities
import gc
import psutil
import os

def check_memory():
    """Check current memory usage"""
    process = psutil.Process(os.getpid())
    memory_mb = process.memory_info().rss / 1024 / 1024
    available_gb = psutil.virtual_memory().available / (1024**3)
    print(f"Current process memory: {memory_mb:.1f} MB")
    print(f"Available system memory: {available_gb:.1f} GB")
    return memory_mb, available_gb

def clear_memory():
    """Force garbage collection to free memory"""
    gc.collect()
    print("Memory cleared!")

# Check initial memory
print("üîç Initial Memory Status:")
check_memory()
````

### 2.3 [Assumptions and Limitations](#24-assumptions-and-limitations)

**Assumptions**
- `salary_in_usd` is assumed to be standardized across all countries regardless of local currency.
- The dataset is considered representative of the global ML job market for the year 2024.
- High-cardinality categorical features such as `job_title` can be effectively captured using embedding layers in a deep learning model.

**Limitations**
- Potential **regional bias** due to overrepresentation from countries like the USA and India.
- The dataset lacks **industry/sector information** (e.g., tech, finance, healthcare), which may impact salary variance.
- **Total compensation** may be underestimated due to missing data on bonuses, equity, or benefits.
- **Temporal limitation**: The dataset only covers the year 2024 and does not account for salary trends over time.

````python
# Cell 1: Import Libraries and Setup
import numpy as np
import pandas as pd

print("üìä STATISTICAL FEATURE ENGINEERING")
print("=" * 50)

# Copy dataset for feature engineering
df_features = data.copy()

# Helper function to display newly created features
def print_new_features(df_before, df_after, section_title):
    new_features = [col for col in df_after.columns if col not in df_before.columns]
    print(f"\nüîπ {section_title} ({len(new_features)} features):")
    for i, feature in enumerate(new_features, 1):
        print(f"{i:2d}. {feature}")

# Cell 2: Salary-Based Statistical Features
print("\nüîÑ Creating Salary-Based Features...")
df_before = df_features.copy()

if 'salary_in_usd' in df_features.columns:
    df_features['salary_log'] = np.log1p(df_features['salary_in_usd'])
    df_features['salary_squared'] = df_features['salary_in_usd'] ** 2
    df_features['salary_bracket'] = pd.cut(
        df_features['salary_in_usd'], 
        bins=[0, 50000, 100000, 150000, 200000, float('inf')], 
        labels=['Low', 'Moderate', 'Medium', 'High', 'Very High']
    )

print_new_features(df_before, df_features, "Salary-Based")
````

## üîß 4.1 Categorical Embeddings for High-Cardinality Variables
___

This subsection addresses the challenge of handling categorical variables with many unique values (high-cardinality). Traditional one-hot encoding becomes impractical with variables like `job_title` and `company_location` that have hundreds of categories. We'll implement embedding layers that learn dense representations of these categorical variables.

````python
# Simplified Categorical Variables Analysis for Embeddings
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Use data DataFrame if it exists, otherwise use df_features
df_to_use = data if 'data' in globals() else df_features

# Get categorical columns
categorical_data = df_to_use.select_dtypes(include=['object', 'category']).columns.tolist()
print(f"Analyzing {len(categorical_data)} categorical variables...")

print("\nüìä CATEGORICAL EMBEDDING RECOMMENDATIONS")
print("=" * 40)

# Store embedding info for presentation
embedding_summary = []

# Analyze each categorical variable
for col in categorical_data:
    unique_vals = df_to_use[col].nunique()
    
    # Simple rule for embedding dimension
    embed_dim = min(50, max(1, unique_vals // 2))
    
    # Decide encoding type
    encoding_type = 'embedding' if unique_vals > 10 else 'one_hot'
    
    # Store information for summary table
    embedding_summary.append({
        'Variable': col,
        'Unique Values': unique_vals,
        'Encoding Type': encoding_type,
        'Embedding Dimension': embed_dim if encoding_type == 'embedding' else 'N/A'
    })

# Create and display summary table
summary_df = pd.DataFrame(embedding_summary)
print(summary_df.to_string(index=False))

# Calculate summary statistics
high_card_vars = sum(1 for item in embedding_summary if item['Encoding Type'] == 'embedding')
low_card_vars = sum(1 for item in embedding_summary if item['Encoding Type'] == 'one_hot')
total_embed_params = sum(item['Unique Values'] * item['Embedding Dimension'] 
                         for item in embedding_summary 
                         if item['Encoding Type'] == 'embedding')

print("\nüìã SUMMARY STATISTICS")
print("=" * 40)
print(f"‚Ä¢ High-cardinality variables (embeddings): {high_card_vars}")
print(f"‚Ä¢ Low-cardinality variables (one-hot): {low_card_vars}")
print(f"‚Ä¢ Total embedding parameters: {total_embed_params:,}")

# Display top embedding variables by parameter count (if any)
if high_card_vars > 0:
    embed_vars = [item for item in embedding_summary if item['Encoding Type'] == 'embedding']
    embed_vars.sort(key=lambda x: x['Unique Values'] * x['Embedding Dimension'], reverse=True)
    
    print("\nüîù TOP 3 VARIABLES BY EMBEDDING SIZE")
    print("=" * 40)
    for i, var in enumerate(embed_vars[:3], 1):
        params = var['Unique Values'] * var['Embedding Dimension']
        print(f"{i}. {var['Variable']}: {params:,} parameters ({var['Unique Values']} x {var['Embedding Dimension']})")
````

````python
# 5.1 Data Preprocessing for Neural Network (Leakage-safe)

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer

# Base dataset (cleaned target without outliers)
base_df = data.copy()

# Ensure correct dtypes
if str(base_df['work_year'].dtype) == 'category':
    base_df['work_year'] = base_df['work_year'].astype(int)

# Define original categorical columns (as strings)
cat_cols = ['employment_type', 'company_size', 'company_location', 'job_title', 
            'employee_residence', 'experience_level', 'remote_ratio']

# Define a helper to create location-based aggregates using TRAIN ONLY
def add_location_aggregates(df_to_transform: pd.DataFrame, stats_source: pd.DataFrame) -> pd.DataFrame:
    stats = stats_source.groupby('company_location')['salary_in_usd'] \
                        .agg(mean='mean', median='median', std='std', count='count')
    out = df_to_transform.copy()
    out['location_avg_salary'] = out['company_location'].map(stats['mean'])
    out['location_salary_std'] = out['company_location'].map(stats['std'])
    out['location_sample_size'] = out['company_location'].map(stats['count'])
    return out

# Split BEFORE computing aggregates or fitting encoders/scalers
train_df, temp_df = train_test_split(base_df, test_size=0.30, random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.50, random_state=42)

# Compute aggregates on TRAIN, map to val/test
train_df = add_location_aggregates(train_df, train_df)
val_df = add_location_aggregates(val_df, train_df)
test_df = add_location_aggregates(test_df, train_df)

# Numeric columns for scaling/imputation
num_cols = ['work_year', 'location_avg_salary', 'location_salary_std', 'location_sample_size']

# Imputers
num_imputer = SimpleImputer(strategy='median')
cat_imputer = SimpleImputer(strategy='most_frequent')

# Impute separately on train, apply to others
train_num = pd.DataFrame(num_imputer.fit_transform(train_df[num_cols]), columns=num_cols)
val_num = pd.DataFrame(num_imputer.transform(val_df[num_cols]), columns=num_cols)
test_num = pd.DataFrame(num_imputer.transform(test_df[num_cols]), columns=num_cols)

train_cat = pd.DataFrame(cat_imputer.fit_transform(train_df[cat_cols]), columns=cat_cols)
val_cat = pd.DataFrame(cat_imputer.transform(val_df[cat_cols]), columns=cat_cols)
test_cat = pd.DataFrame(cat_imputer.transform(test_df[cat_cols]), columns=cat_cols)

# Ordinal encode categoricals with unknowns handled
ord_enc = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
train_cat_enc = pd.DataFrame(ord_enc.fit_transform(train_cat), columns=cat_cols)
val_cat_enc = pd.DataFrame(ord_enc.transform(val_cat), columns=cat_cols)
test_cat_enc = pd.DataFrame(ord_enc.transform(test_cat), columns=cat_cols)

# Scale numeric features (fit on train only)
scaler = StandardScaler()
train_num_s = pd.DataFrame(scaler.fit_transform(train_num), columns=num_cols)
val_num_s = pd.DataFrame(scaler.transform(val_num), columns=num_cols)
test_num_s = pd.DataFrame(scaler.transform(test_num), columns=num_cols)

# Combine encoded categorical and scaled numeric features
X_train = pd.concat([train_cat_enc.reset_index(drop=True), train_num_s.reset_index(drop=True)], axis=1)
X_val = pd.concat([val_cat_enc.reset_index(drop=True), val_num_s.reset_index(drop=True)], axis=1)
X_test = pd.concat([test_cat_enc.reset_index(drop=True), test_num_s.reset_index(drop=True)], axis=1)

# Targets
y_train = train_df['salary_in_usd'].reset_index(drop=True)
y_val = val_df['salary_in_usd'].reset_index(drop=True)
y_test = test_df['salary_in_usd'].reset_index(drop=True)

# Keep transformers for inference
encoders = { 'ordinal_encoder': ord_enc }
num_scaler = scaler

print(f"Training set: {X_train.shape[0]} samples")
print(f"Validation set: {X_val.shape[0]} samples")
print(f"Test set: {X_test.shape[0]} samples")
print(f"Features: {X_train.shape[1]} columns")
print("\nCategorical features encoded (train classes only; unknowns handled as -1):")
for col, n in zip(cat_cols, [len(c) if hasattr(encoders['ordinal_encoder'].categories_[i], '__len__') else 'N/A' 
                             for i, c in enumerate(encoders['ordinal_encoder'].categories_)]):
    print(f"- {col}: {n} classes")
print("\nNumerical features scaled:")
for col in num_cols:
    print(f"- {col}")
````

For categorical variables with high cardinality (many unique values), such as job titles and company locations, embeddings offer significant advantages over traditional one-hot encoding:

- **Semantic Relationships**: Captures relationships between categorical values (similar job titles cluster together)

````python
# 5.2 PyTorch Neural Network Implementation (with scaled target, features already preprocessed)

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
from sklearn.preprocessing import StandardScaler

# Set device
torch.manual_seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

# Scale the target variable ONLY for numerical stability
_target_scaler = StandardScaler()
y_train_scaled = _target_scaler.fit_transform(y_train.values.reshape(-1, 1)).flatten()
y_val_scaled = _target_scaler.transform(y_val.values.reshape(-1, 1)).flatten()
y_test_scaled = _target_scaler.transform(y_test.values.reshape(-1, 1)).flatten()

print(f"Target scaling - Mean: {_target_scaler.mean_[0]:.2f}, Std: {_target_scaler.scale_[0]:.2f}")

# Convert data to tensors
X_train_tensor = torch.FloatTensor(X_train.values).to(device)
y_train_tensor = torch.FloatTensor(y_train_scaled).reshape(-1, 1).to(device)
X_val_tensor = torch.FloatTensor(X_val.values).to(device)
y_val_tensor = torch.FloatTensor(y_val_scaled).reshape(-1, 1).to(device)
X_test_tensor = torch.FloatTensor(X_test.values).to(device)
y_test_tensor = torch.FloatTensor(y_test_scaled).reshape(-1, 1).to(device)

# Data loaders
batch_size = 64
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
test_dataset = TensorDataset(X_test_tensor, y_test_tensor)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Simple MLP baseline (features already include encoded categoricals + scaled numerics)
class SimpleSalaryPredictor(nn.Module):
    def __init__(self, input_dim):
        super(SimpleSalaryPredictor, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 1)
        )
    def forward(self, x):
        return self.model(x)

input_dim = X_train.shape[1]
model = SimpleSalaryPredictor(input_dim).to(device)
print(model)

criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)

target_scaler = _target_scaler  # keep reference for evaluation
````

## 5.4 Hyperparameter Tuning

Neural networks require careful tuning of hyperparameters to achieve optimal performance. Key hyperparameters include:

1. **Learning Rate**: Controls how quickly the model adapts to the problem (typically 0.001-0.01)
2. **Batch Size**: Number of samples processed before updating model parameters (32, 64, 128, 256)
3. **Number of Hidden Layers**: Depth of the network (typically 2-5 for this task)
4. **Neurons per Layer**: Width of the network (decreasing pattern from input to output)
5. **Dropout Rate**: Regularization to prevent overfitting (0.1-0.5)
6. **Embedding Dimensions**: Size of embeddings for categorical variables

For systematic tuning, we use Optuna, which implements Bayesian optimization to efficiently search through the hyperparameter space.

````python
# 5.3 Model Training

import os
import time
import matplotlib.pyplot as plt

num_epochs = 100
best_val_loss = float('inf')
early_stopping_patience = 15
early_stopping_counter = 0

train_losses = []
val_losses = []

model_dir = "model_checkpoints"
os.makedirs(model_dir, exist_ok=True)
model_path = os.path.join(model_dir, "best_model.pth")

print(f"Starting training with scaled target...")
start_time = time.time()

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, targets in train_loader:
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        running_loss += loss.item() * inputs.size(0)

    epoch_train_loss = running_loss / len(train_dataset)
    train_losses.append(epoch_train_loss)

    model.eval()
    running_val_loss = 0.0
    with torch.no_grad():
        for inputs, targets in val_loader:
            outputs = model(inputs)
            val_loss = criterion(outputs, targets)
            running_val_loss += val_loss.item() * inputs.size(0)

    epoch_val_loss = running_val_loss / len(val_dataset)
    val_losses.append(epoch_val_loss)

    if (epoch + 1) % 10 == 0 or epoch == 0:
        print(f'Epoch {epoch+1}/{num_epochs}, Train Loss: {epoch_train_loss:.4f}, Val Loss: {epoch_val_loss:.4f}')

    if epoch_val_loss < best_val_loss:
        best_val_loss = epoch_val_loss
        early_stopping_counter = 0
        torch.save(model.state_dict(), model_path)
        print(f"Model saved at epoch {epoch+1} with validation loss: {epoch_val_loss:.4f}")
    else:
        early_stopping_counter += 1
        if early_stopping_counter >= early_stopping_patience:
            print(f'Early stopping at epoch {epoch+1}')
            break

training_time = time.time() - start_time
print(f"Training completed in {training_time/60:.2f} minutes")

if os.path.exists(model_path):
    model.load_state_dict(torch.load(model_path))
    print(f"Loaded best model from {model_path}")

plt.figure(figsize=(10, 6))
plt.plot(train_losses, label='Training Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.title('Training and Validation Loss')
plt.legend()
plt.grid(True)
plt.show()
````

````python
# 5.7 Hyperparameter Tuning with Optuna

# Install Optuna if not already installed
%pip install optuna

import optuna
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score

# Define the objective function for Optuna
def objective(trial):
    # Define the hyperparameters to tune
    hidden_size_1 = trial.suggest_int('hidden_size_1', 32, 256, step=32)
    hidden_size_2 = trial.suggest_int('hidden_size_2', 16, 128, step=16)
    dropout_rate = trial.suggest_float('dropout_rate', 0.1, 0.5, step=0.1)
    learning_rate = trial.suggest_float('learning_rate', 1e-4, 1e-2, log=True)
    batch_size = trial.suggest_categorical('batch_size', [16, 32, 64, 128])
    
    # Create model with the suggested hyperparameters
    class TunedModel(nn.Module):
        def __init__(self):
            super(TunedModel, self).__init__()
            self.fc1 = nn.Linear(input_dim, hidden_size_1)
            self.dropout1 = nn.Dropout(dropout_rate)
            self.fc2 = nn.Linear(hidden_size_1, hidden_size_2)
            self.dropout2 = nn.Dropout(dropout_rate)
            self.fc3 = nn.Linear(hidden_size_2, 1)
            
        def forward(self, x):
            x = torch.relu(self.fc1(x))
            x = self.dropout1(x)
            x = torch.relu(self.fc2(x))
            x = self.dropout2(x)
            return self.fc3(x)
    
    # Create model
    model = TunedModel().to(device)
    
    # Define loss function and optimizer
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    
    # Create data loaders with the suggested batch size
    train_dataset = torch.utils.data.TensorDataset(X_train_tensor, y_train_tensor)
    val_dataset = torch.utils.data.TensorDataset(X_val_tensor, y_val_tensor)
    
    train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=batch_size)
    
    # Training loop with early stopping
    best_val_loss = float('inf')
    early_stopping_counter = 0
    early_stopping_patience = 5
    
    for epoch in range(30):  # Limit to 30 epochs for tuning
        # Training
        model.train()
        running_loss = 0.0
        
        for inputs, targets in train_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            
            # Zero the parameter gradients
            optimizer.zero_grad()
            
            # Forward pass
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            
            # Backward pass and optimize
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item() * inputs.size(0)
        
        # Validation
        model.eval()
        val_loss = 0.0
        
        with torch.no_grad():
            for inputs, targets in val_loader:
                inputs, targets = inputs.to(device), targets.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, targets)
                val_loss += loss.item() * inputs.size(0)
        
        val_loss /= len(val_loader.dataset)
        
        # Early stopping check
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            early_stopping_counter = 0
        else:
            early_stopping_counter += 1
            if early_stopping_counter >= early_stopping_patience:
                break
    
    # Return the best validation loss as the objective to minimize
    return best_val_loss

# Run the hyperparameter tuning
print("Starting hyperparameter optimization with Optuna...")
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=10)  # Adjust n_trials as needed

# Print the best parameters and value
print('Best trial:')
trial = study.best_trial
print('  Value (Best Validation Loss): {}'.format(trial.value))
print('  Params: ')
for key, value in trial.params.items():
    print('    {}: {}'.format(key, value))

# Create a model with the best hyperparameters
best_hidden_size_1 = trial.params['hidden_size_1']
best_hidden_size_2 = trial.params['hidden_size_2']
best_dropout_rate = trial.params['dropout_rate']
best_learning_rate = trial.params['learning_rate']
best_batch_size = trial.params['batch_size']

print("\nTraining final model with best hyperparameters...")

class OptimizedModel(nn.Module):
    def __init__(self):
        super(OptimizedModel, self).__init__()
        self.fc1 = nn.Linear(input_dim, best_hidden_size_1)
        self.dropout1 = nn.Dropout(best_dropout_rate)
        self.fc2 = nn.Linear(best_hidden_size_1, best_hidden_size_2)
        self.dropout2 = nn.Dropout(best_dropout_rate)
        self.fc3 = nn.Linear(best_hidden_size_2, 1)
        
    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.dropout1(x)
        x = torch.relu(self.fc2(x))
        x = self.dropout2(x)
        return self.fc3(x)

# Create and train the optimized model
optimized_model = OptimizedModel().to(device)
optimizer = optim.Adam(optimized_model.parameters(), lr=best_learning_rate)
criterion = nn.MSELoss()

# Create data loaders with the best batch size
train_dataset = torch.utils.data.TensorDataset(X_train_tensor, y_train_tensor)
val_dataset = torch.utils.data.TensorDataset(X_val_tensor, y_val_tensor)
test_dataset = torch.utils.data.TensorDataset(X_test_tensor, y_test_tensor)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=best_batch_size, shuffle=True)
val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=best_batch_size)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=best_batch_size)

# Training loop
best_val_loss = float('inf')
early_stopping_counter = 0
early_stopping_patience = 10
train_losses = []
val_losses = []
best_model_path = 'optimized_model.pth'

import time
start_time = time.time()

for epoch in range(100):  # Increased max epochs for final training
    # Training
    optimized_model.train()
    running_loss = 0.0
    
    for inputs, targets in train_loader:
        inputs, targets = inputs.to(device), targets.to(device)
        
        # Zero the parameter gradients
        optimizer.zero_grad()
        
        # Forward pass
        outputs = optimized_model(inputs)
        loss = criterion(outputs, targets)
        
        # Backward pass and optimize
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item() * inputs.size(0)
    
    epoch_train_loss = running_loss / len(train_loader.dataset)
    train_losses.append(epoch_train_loss)
    
    # Validation
    optimized_model.eval()
    running_val_loss = 0.0
    
    with torch.no_grad():
        for inputs, targets in val_loader:
            inputs, targets = inputs.to(device), targets.to(device)
            outputs = optimized_model(inputs)
            loss = criterion(outputs, targets)
            running_val_loss += loss.item() * inputs.size(0)
    
    epoch_val_loss = running_val_loss / len(val_loader.dataset)
    val_losses.append(epoch_val_loss)
    
    print(f'Epoch {epoch+1}, Training Loss: {epoch_train_loss:.4f}, Validation Loss: {epoch_val_loss:.4f}')
    
    # Early stopping check
    if epoch_val_loss < best_val_loss:
        best_val_loss = epoch_val_loss
        torch.save(optimized_model.state_dict(), best_model_path)
        early_stopping_counter = 0
    else:
        early_stopping_counter += 1
        if early_stopping_counter >= early_stopping_patience:
            print(f'Early stopping triggered after {epoch+1} epochs')
            break

optimized_training_time = time.time() - start_time
print(f'Training completed in {optimized_training_time:.2f} seconds')

# Load the best model
optimized_model.load_state_dict(torch.load(best_model_path))

# Evaluate on test set
optimized_model.eval()
y_pred_list = []
y_true_list = []

with torch.no_grad():
    for inputs, targets in test_loader:
        inputs, targets = inputs.to(device), targets.to(device)
        outputs = optimized_model(inputs)
        y_pred_list.extend(outputs.cpu().numpy())
        y_true_list.extend(targets.cpu().numpy())

y_pred_optimized = np.array(y_pred_list).flatten()
y_true = np.array(y_true_list).flatten()

# Calculate metrics
optimized_mse = mean_squared_error(y_true, y_pred_optimized)
optimized_rmse = np.sqrt(optimized_mse)
optimized_r2 = r2_score(y_true, y_pred_optimized)

print("\nTest Metrics for Optimized Model:")
print(f"MSE: {optimized_mse:.4f}")
print(f"RMSE: {optimized_rmse:.4f}")
print(f"R¬≤: {optimized_r2:.4f}")

# Compare with previous model
print("\nComparison with Original Model:")
print(f"Original Model - MSE: {mse:.4f}, R¬≤: {r2:.4f}")
print(f"Optimized Model - MSE: {optimized_mse:.4f}, R¬≤: {optimized_r2:.4f}")
improvement = ((mse - optimized_mse) / mse) * 100
print(f"Improvement in MSE: {improvement:.2f}%")

# Log the optimized model with MLflow
import mlflow
import mlflow.pytorch

# Start a new MLflow run
with mlflow.start_run(run_name="optimized_pytorch_nn"):
    # Log hyperparameters
    for key, value in trial.params.items():
        mlflow.log_param(key, value)
    
    # Log metrics
    mlflow.log_metric("optimized_mse", optimized_mse)
    mlflow.log_metric("optimized_rmse", optimized_rmse)
    mlflow.log_metric("optimized_r2", optimized_r2)
    mlflow.log_metric("improvement_percentage", improvement)
    mlflow.log_metric("optimized_training_time", optimized_training_time)
    
    # Log the PyTorch model
    mlflow.pytorch.log_model(optimized_model, "optimized_pytorch_model")
    
    # Log training and validation loss history
    for epoch_idx, (train_loss, val_loss) in enumerate(zip(train_losses, val_losses)):
        mlflow.log_metric("optimized_train_loss", train_loss, step=epoch_idx)
        mlflow.log_metric("optimized_val_loss", val_loss, step=epoch_idx)
    
    print(f"MLflow tracking completed for optimized model. Run ID: {mlflow.active_run().info.run_id}")
    print(f"Optimized model artifacts saved at: {mlflow.get_artifact_uri()}")

# Visualize training and validation loss
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(train_losses, label='Training Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.title('Training and Validation Loss for Optimized Model')
plt.legend()
plt.grid(True)
plt.show()

# Visualize predictions vs actual values
plt.figure(figsize=(10, 6))
plt.scatter(y_true, y_pred_optimized, alpha=0.5)
plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--')
plt.xlabel('Actual Salary')
plt.ylabel('Predicted Salary')
plt.title('Actual vs Predicted Salary (Optimized Model)')
plt.grid(True)
plt.show()
````

````python
# 5.7 Hyperparameter Tuning with Optuna

import optuna
from optuna.integration.mlflow import MLflowCallback

# Define objective function for Optuna
def objective(trial):
    # Hyperparameters to optimize
    lr = trial.suggest_float("learning_rate", 1e-4, 1e-2, log=True)
    hidden1 = trial.suggest_int("hidden_layer1", 32, 256)
    hidden2 = trial.suggest_int("hidden_layer2", 16, 128)
    dropout1 = trial.suggest_float("dropout1", 0.1, 0.5)
    dropout2 = trial.suggest_float("dropout2", 0.1, 0.5)
    batch_size = trial.suggest_categorical("batch_size", [16, 32, 64, 128])
    
    # Create model with trial hyperparameters
    class TrialModel(nn.Module):
        def __init__(self):
            super(TrialModel, self).__init__()
            self.model = nn.Sequential(
                nn.Linear(input_dim, hidden1),
                nn.ReLU(),
                nn.Dropout(dropout1),
                nn.Linear(hidden1, hidden2),
                nn.ReLU(),
                nn.Dropout(dropout2),
                nn.Linear(hidden2, 1)
            )
        
        def forward(self, x):
            return self.model(x)
    
    # Initialize the model
    trial_model = TrialModel().to(device)
    
    # Loss function and optimizer
    trial_criterion = nn.MSELoss()
    trial_optimizer = optim.Adam(trial_model.parameters(), lr=lr)
    
    # DataLoaders
    trial_train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    trial_val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    
    # Training
    num_epochs = 50  # Reduced for faster tuning
    best_val_loss = float('inf')
    early_stopping_patience = 5  # Reduced for faster tuning
    early_stopping_counter = 0
    
    for epoch in range(num_epochs):
        # Training phase
        trial_model.train()
        for inputs, targets in trial_train_loader:
            outputs = trial_model(inputs)
            loss = trial_criterion(outputs, targets)
            
            trial_optimizer.zero_grad()
            loss.backward()
            trial_optimizer.step()
        
        # Validation phase
        trial_model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for inputs, targets in trial_val_loader:
                outputs = trial_model(inputs)
                val_loss += trial_criterion(outputs, targets).item() * inputs.size(0)
        
        val_loss /= len(val_dataset)
        
        # Early stopping
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            early_stopping_counter = 0
        else:
            early_stopping_counter += 1
            if early_stopping_counter >= early_stopping_patience:
                break
    
    # Return best validation loss
    return best_val_loss

# MLflow callback
mlflow_callback = MLflowCallback(
    tracking_uri=mlflow.get_tracking_uri(),
    metric_name="val_loss"
)

# Create a study object and optimize
print("Starting hyperparameter tuning...")
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=20, callbacks=[mlflow_callback])

# Print results
print("Best trial:")
trial = study.best_trial
print(f"  Value: {trial.value}")
print("  Params: ")
for key, value in trial.params.items():
    print(f"    {key}: {value}")

# Train final model with best parameters
best_params = study.best_params

# Create model with best hyperparameters
class BestModel(nn.Module):
    def __init__(self):
        super(BestModel, self).__init__()
        self.model = nn.Sequential(
            nn.Linear(input_dim, best_params["hidden_layer1"]),
            nn.ReLU(),
            nn.Dropout(best_params["dropout1"]),
            nn.Linear(best_params["hidden_layer1"], best_params["hidden_layer2"]),
            nn.ReLU(),
            nn.Dropout(best_params["dropout2"]),
            nn.Linear(best_params["hidden_layer2"], 1)
        )
    
    def forward(self, x):
        return self.model(x)

# Final model training
final_model = BestModel().to(device)
print("Final model architecture:")
print(final_model)
````

````python
# 5.8 Final Model Training and Evaluation

# Training the final model with best hyperparameters
criterion = nn.MSELoss()
optimizer = optim.Adam(final_model.parameters(), lr=best_params["learning_rate"])
batch_size = best_params["batch_size"]

# Create data loaders with best batch size
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Training parameters
num_epochs = 100
best_val_loss = float('inf')
early_stopping_patience = 10
early_stopping_counter = 0

# Lists to track metrics
train_losses = []
val_losses = []

# Training loop
print("Training final model with best hyperparameters...")
for epoch in range(num_epochs):
    # Training phase
    final_model.train()
    running_loss = 0.0
    for inputs, targets in train_loader:
        outputs = final_model(inputs)
        loss = criterion(outputs, targets)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        running_loss += loss.item() * inputs.size(0)
    
    epoch_train_loss = running_loss / len(train_dataset)
    train_losses.append(epoch_train_loss)
    
    # Validation phase
    final_model.eval()
    running_val_loss = 0.0
    with torch.no_grad():
        for inputs, targets in val_loader:
            outputs = final_model(inputs)
            val_loss = criterion(outputs, targets)
            running_val_loss += val_loss.item() * inputs.size(0)
    
    epoch_val_loss = running_val_loss / len(val_dataset)
    val_losses.append(epoch_val_loss)
    
    # Print progress
    if (epoch + 1) % 10 == 0:
        print(f'Epoch {epoch+1}/{num_epochs}, '
              f'Train Loss: {epoch_train_loss:.4f}, '
              f'Val Loss: {epoch_val_loss:.4f}')
    
    # Early stopping
    if epoch_val_loss < best_val_loss:
        best_val_loss = epoch_val_loss
        early_stopping_counter = 0
        # Save the best model
        torch.save(final_model.state_dict(), 'best_final_model.pth')
    else:
        early_stopping_counter += 1
        if early_stopping_counter >= early_stopping_patience:
            print(f'Early stopping at epoch {epoch+1}')
            break

# Load the best model
final_model.load_state_dict(torch.load('best_final_model.pth'))

# Plot training and validation loss
plt.figure(figsize=(10, 6))
plt.plot(train_losses, label='Training Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Final Model Training and Validation Loss')
plt.legend()
plt.grid(True)
plt.show()

# Evaluate on test set
final_model.eval()
all_outputs = []
all_targets = []

with torch.no_grad():
    for inputs, targets in test_loader:
        outputs = final_model(inputs)
        all_outputs.append(outputs.cpu().numpy())
        all_targets.append(targets.cpu().numpy())

# Convert lists to numpy arrays
y_pred = np.vstack(all_outputs).flatten()
y_true = np.vstack(all_targets).flatten()

# Calculate evaluation metrics
mse = mean_squared_error(y_true, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_true, y_pred)
r2 = r2_score(y_true, y_pred)

print("\nFinal Model Evaluation Metrics:")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R¬≤ Score: {r2:.4f}")

# Create a scatter plot of predicted vs actual values
plt.figure(figsize=(10, 8))
plt.scatter(y_true, y_pred, alpha=0.5)
plt.plot([y_true.min(), y_true.max()], [y_true.min(), y_true.max()], 'r--')
plt.xlabel('Actual Salary')
plt.ylabel('Predicted Salary')
plt.title('Final Model: Actual vs Predicted Salary')
plt.grid(True)
plt.show()
````
